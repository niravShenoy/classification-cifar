```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import itertools
```

```python
def loss_function(param):
#     plots = ['']
    plots = ['-60','-75','-95']
    file= param + '.csv'
    foldername = os.path.join(curr_path, 'Task1_pseudoLabeling', 'logs')
    all_loss = []
    all_accuracy = []
    colors = ['b', 'r', 'g', 'c']
    cc = itertools.cycle(colors)
    for plot in plots:
        loss_list = []
        filepath = os.path.join(foldername, param, param + plot + '.csv')
        df = pd.read_csv(filepath)
        c = next(cc)
        for xiter, x in enumerate(df['tloss']):

            if (np.isnan(x) == False):
                loss_list.append(x)
        all_loss.append(loss_list)
        plt.plot(loss_list, label=param+plot)
#         all_loss.append(l1)
    plt.title(param + ' Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend(loc='upper right')
    plt.savefig(file.split('.')[0] + 'loss'+'.png')
    plt.show()
        
    for plot in plots:
        accuracy_list = []
        filepath = os.path.join(foldername, param, param + plot + '.csv')
        df = pd.read_csv(filepath)
        c = next(cc)
        for xiter, x in enumerate(df['taccuracy']):
            if (np.isnan(x) == False):
                accuracy_list.append(x)
        all_accuracy.append(accuracy_list)
    #     plt.show()
    #     fig.savefig(file.split('.')[0] + 'loss'+'.png', dpi=fig.dpi)
        plt.plot(accuracy_list, label=param+plot)
    plt.title(param + ' Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend(loc='lower right')
    plt.savefig(file.split('.')[0] + 'accuracy''.png')
    plt.show()
```
